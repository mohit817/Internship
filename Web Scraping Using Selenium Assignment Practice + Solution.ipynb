```python
# Question 4
# Scrape data of first 100 sunglasses listings on flipkart.com
```

```python
driver.get('http://www.flipkart.com/')
```

```python
search_button = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input') 
search_button.send_keys('sunglasses')
```

```python
click_search_button = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button')
click_search_button.submit()
```

```python
brand_name = []
brand_description = []
brand_price = []
brand_discount = []
```

```python
while len(brand_name) <= 100:
    name_tags = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    description_tags = driver.find_elements(By.CLASS_NAME,"IRpwTa")
    price_tags = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    discount_tags = driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    
    for i in name_tags:
        name = i.text
        brand_name.append(name)
    
    for i in description_tags:
        description = i.text
        brand_description.append(description)
    
    for i in price_tags:
        price = i.text
        brand_price.append(price)
    
    for i in discount_tags:
        discount = i.text
        brand_discount.append(discount)
        
    next_page = driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_page.click()
    time.sleep(3)
```

```python
print(len(brand_name),
len(brand_description),
len(brand_price),
len(brand_discount))
```

```python
df = pd.DataFrame({'Name':brand_name,'Description':brand_description,'Price':brand_price,'Discount':brand_discount})
df.head(100)
```

```python
# Question 8
# Write a python program to scrape data for Top 1000 Quotes of All Time.
# The above task will be done in following steps:
# 1. First get the webpagehttps://www.azquotes.com/
# 2. Click on TopQuotes
# 3. Than scrap a) Quote b) Author c) Type Of Quotes
```

```python
driver.get('http://www.azquotes.com/')
```

```python
top_quotes_click = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top_quotes_click.click()
```

```python
quote = []
author = []
type_quote = []
```

```python
num_pages = 10
base_url = "https://www.azquotes.com/top_quotes.html?p="
for page in range(1,num_pages + 1):
    quote_tags = driver.find_elements(By.XPATH,'//a[@class="title"]')
    author_tags = driver.find_elements(By.XPATH,'//div[@class="author"]')
    type_quote_tags = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in quote_tags:
        quote.append(i.text)
    for i in author_tags:
        author.append(i.text)
    for i in type_quote_tags:
        type_quote.append(i.text)
    driver.get(base_url + str(page))
    time.sleep(3)
```

```python
df = pd.DataFrame({'Quote':quote,'Author':author,'Type of Quote':type_quote})
df
```
